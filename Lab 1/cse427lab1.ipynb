# Numpy Library tasks

```python
#1
import numpy as np
a=np.random.randint(1,101,20)
b=np.random.randint(1,101,20)
```

```python
#2
print(f"Old shape A: {a.shape}")
print(f"Old shape B: {b.shape}")
a=a.reshape(5,4)
b=b.reshape(5,4)
print(f"New shape A: {a.shape}")
print(f"New shape B: {b.shape}")
print(a)
print()
print(b)
```

```python
#3
scaler=int(input("Enter a val : "))
a = a*scaler
b = b*scaler
print(a)
print()
print(b)
```

```python
#4
b=b.T
c=a@b
print(c)
```

```python
#5
print(c[:,1:3])
print(c[:])
```

```python
#6
max_value = np.max(c)
max_index = np.unravel_index(np.argmax(c), c.shape)
print("\nLargest element in c:", max_value)
print("Index of largest element:", max_index)
```

```python
print("original array:",c)
d = c.flatten() 
print("--------------")
print("--------------")
print("converted array:",d)
```

# Matplotlib Tasks

```python
#1
import matplotlib.pyplot as plt
temperature_dhaka = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_dhaka = [28, 25, 29, 20, 26, 50, 19, 29, 52, 55]
plt.title("temp vs humidity")
plt.xlabel("temp")
plt.ylabel("humidity")
plt.grid()
plt.scatter(temperature_dhaka, humidity_dhaka, color="red", marker="*")
plt.show()
```

```python
#2
study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]
plt.figure(figsize=(12, 8))
plt.plot(study_hours, marks, marker='o', linestyle='-', color='black', linewidth=2)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.grid(True)
plt.show()
```

```python
#3
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]
plt.figure(figsize=(12, 6))
plt.barh(subjects, marks, color='skyblue')
plt.xlabel("marks")
plt.ylabel("Subjects")
plt.title("marks in Each Subject")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
#4
colors = ['red', 'blue','purple', 'cyan', 'magenta']
plt.figure(figsize=(12,6))
plt.bar(subjects, marks, color=colors)
plt.xlabel("Subjects")
plt.ylabel("marks")
plt.title("marks in Each Subject")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#5

subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]

plt.subplot(2, 2, 3)
plt.barh(subjects, marks, color='skyblue')
plt.xlabel("Marks")
plt.ylabel("Subjects")
plt.title("Marks in Each Subject")
plt.grid(axis='x', linestyle='--', alpha=0.7)


plt.subplot(2, 2, 4)
plt.bar(subjects, marks, color=colors)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Marks in Each Subject (Colored)")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()


plt.show()
```
